```python
#objective-To predict whether the given person survived in titanic mishap or not.
```

## Data Ingestion

```python
import seaborn as sns
```

```python
titanic=sns.load_dataset("titanic")
```

```python
titanic
```

## Feature Extraction and EDA

```python
titanic.info()
```

```python
# Analysing if Priority Class had any impact on Survival Rate
```

```python
titanic[["pclass","survived"]].groupby(["pclass"],as_index=False).mean()
```

```python
# Insight
# from the above analysis we have observed that passenger having higher class have more chance to be survived
```

```python
# Analyse Wheather there is any realation between the family size and the survival rate
```

```python

# sibsp = no of sibling/spouses aboard
# parch = no of parent/children aboard
titanic["familysize"]=titanic["sibsp"]+titanic["parch"]+1
titanic[["familysize","survived"]].groupby("familysize",as_index=False).mean()
```

```python
# insight
# from the above analysis we have observed that the moderate family size have better chance for survival
```

```python
# Lets analyze wheather gender get impacted for survival or not
titanic[["sex","survived"]].groupby("sex",as_index=False).mean()
```

```python
#insight
# from the above output we have observed that female passengers have the maximum chance to be survived
```

```python
# Lets Analyse wheather the base of city get impacted for survival or not
```

```python
titanic[["embarked","survived"]].groupby("embarked",as_index=False).mean()
```

```python
# from the above analysis we have observed that The town of "C" has the maximum chance to be survived
```

```python

##### Analysing if Fare paid by passengers had any impact on Survival Rate
```

```python
titanic[["fare","survived"]].groupby("fare",as_index=False).mean()
```

```python
# since the above analysis is to complecated to understand we have to add a categorical column for the fare 
```

```python
titanic["fare"]=titanic["fare"].fillna(titanic["fare"].median())
```

```python
titanic["categoricalfare"]=pd.qcut(titanic["fare"],3)
```

```python
titanic[["categoricalfare","survived"]].groupby("categoricalfare",as_index=False).mean()
```

```python
# from the above analysis we have observed that those passengers who have paid maximum fare have better chance of survival
```

```python
titanic.head()
```

```python

##### Analysing if Age of the passengers had any impact on Survival Rate
```

```python
# We have earlier seen the Age feature has considerable nulls. Hence we can impute the nulls before categorising Age and analysing impact on survival.
```

```python
age_avg=titanic["age"].mean()
age_std=titanic["age"].std()
age_null_counts=titanic["age"].isnull().sum()
```

```python
#random age imputation so that data is complete
#this process should only be done after problem consideration, its not a std process
```

```python
age_null_random_list = np.random.randint(age_avg - 3*age_std, age_avg + 3*age_std, size=age_null_counts)
```

```python
#can take 0 as min also
titanic['age'][np.isnan(titanic['age'])] = age_null_random_list
titanic['age'] = titanic['age'].astype(int)
```

```python
# making a catagorical field
titanic["categoricalage"]=pd.cut(titanic["age"],3)
```

```python
titanic[["categoricalage","survived"]].groupby("categoricalage",as_index=False).mean()
```

```python
# from the above analysis we have observed that the peoples under 18 have the maximum chances of suirvival
```

```python

#####  Assigning labels to extracted categorical variables - Label encoding
```

```python
# for sex
titanic["sex"]=titanic["sex"].map({"male":1,"female":1}).astype(int)
```

```python
# for embarked
titanic["embarked"][titanic.embarked=="S"]=0
titanic["embarked"][titanic.embarked=="C"]=1
titanic["embarked"][titanic.embarked=="Q"]=2
```

```python
# for fare
titanic["fare"][titanic.fare<=7.91]=0
titanic["fare"][(titanic.fare<7.91)&(titanic.fare>=14.454)]=1
titanic["fare"][(titanic.fare<14.454)&(titanic.fare>=31)]=2
titanic["fare"][titanic.fare>31]=3
titanic["fare"]=titanic["fare"].astype(int)
```

```python
# for age
titanic["age"][titanic.age<=16]=0
titanic["age"][(titanic.age>16)&(titanic.age>=32)]=1
titanic["age"][(titanic.age>32)&(titanic.age>=48)]=2
titanic["age"][(titanic.age>48)&(titanic.age>=64)]=3
titanic["age"][titanic.age>64]=4
```

### Feature selection

```python
drop_elements=['sibsp', 'parch', 'categoricalage', 'categoricalfare']
```

```python
titanic=titanic.drop(drop_elements,axis=1)
```

```python
titanic.head(10)
```
